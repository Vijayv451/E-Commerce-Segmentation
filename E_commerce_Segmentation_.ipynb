



# SCALING

```python
x = df1.iloc[:,2:].values
x
```

```python
scale = MinMaxScaler()
features = scale.fit_transform(x)
features
```

# ELBOW METHOD TO FIND K-VALUE

```python
inertia = []
for i in range(1, 16):
    k_means = KMeans(n_clusters=i, n_init=10)  # Explicitly set the value of n_init
    k_means = k_means.fit(features)
    inertia.append(k_means.inertia_)
```

```python
# elbow graph
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.plot(range(1,16), inertia, 'bo-')
plt.xlabel('No of clusters'), plt.ylabel('Inertia')
```

```python
# Create a K-means model and set n_init explicitly
model = KMeans(random_state=42, n_init=10)

# Create the KElbowVisualizer
visualizer = KElbowVisualizer(model, k=(1, 11))

# Fit the visualizer to the data
visualizer.fit(features)
plt.suptitle("Elbow Graph and Elbow Visualizer")

# Visualize the elbow plot
visualizer.show()
```

# Silhoutte Score for k values

```python
silhouette_avg = []
k_values = range(2, 16)

for k in k_values:
    # Create a K-means model and set n_init explicitly
    kmeans = KMeans(n_clusters=k, n_init=10)
    cluster_labels = kmeans.fit_predict(features)
    silhouette_avg.append(silhouette_score(features, cluster_labels))
```

```python
import warnings

# Suppress the warning
warnings.filterwarnings('ignore', category=FutureWarning)

silhouette_avg = []
k_values = range(2, 16)

for k in k_values:
    # Create a K-means model and set n_init explicitly
    kmeans = KMeans(n_clusters=k, n_init=10)
    cluster_labels = kmeans.fit_predict(features)
    silhouette_avg.append(silhouette_score(features, cluster_labels))

plt.figure(figsize=(10, 7))
plt.plot(k_values, silhouette_avg, 'bX-')
plt.xlabel('Values of K')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis for Optimal K')
plt.show()
```

```python
model = KMeans(n_clusters=3)
model = model.fit(features)
y_km = model.predict(features)
centers = model.cluster_centers_
df1['Cluster'] = pd.DataFrame(y_km)
df1.to_csv("Cluster_data",index=False)
df1['Cluster'].value_counts()
```
